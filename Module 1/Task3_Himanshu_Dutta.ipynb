```python
#ALl the unique values in the dataframe
columns = data.columns.unique().tolist()
unique_val = []
for i in columns:
    unique_val.append(pd.unique(data[columns].values.ravel('k')).tolist())
import pprint
# pprint.pprint(unique_val)
```

```python
#Converting the date_time column into pandas date_time format and saving it into a new row
data['new_date_time'] = pd.to_datetime(data['date_time'], format = '%Y-%m-%d %H:%M:%S.%f')
data['new_date_time']
```

```python
#Checking for duplicate rows
duplicate_rows = data[data.duplicated()]
duplicate_rows
```

```python
#Checking for duplicate column
duplicate_columns = data.T[data.T.duplicated()]
#droping the duplicate column
data = data.T.drop(duplicate_columns.T.columns).T
duplicate_columns
```

```python
#Finding the constant columns
const_col = []
for i in data.columns:
    if len(data[i].unique()) == 1:
        const_col.append(i)
const_col
```

```python
data_dict = {
    'columns': data.columns.tolist(),
    'data_type' : data.dtypes.tolist(),
    'describe' : data.describe(),
    'num_cat' : np.issubdtype(data[data.columns].dtypes, np.number)
}

df = pd.DataFrame(data=data_dict)

df = (df.T)

print (df)

df.to_excel('dict1.xlsx')
```
