```python
#Numerical Data in the dataframe
data_numeric = data.select_dtypes(include=['int64', 'float64'])
data_numeric.head()
```

```python
#Checking for correlation numerically
data_corr = data_numeric.corr()['SalePrice'][:]
data_corr
```

The correlation of **SalePrice** with other numeric values are shown above. We require the correlation to be as high as possible, i.e, as close to 1 or -1 as possible. This shows a strong relation b/w the two variables, 0 signifying a weak relation, which we don't need.

```python
#Plotting PairPlot between various input features and the target variable
for i in range(1,len(data_numeric.columns),5):
    sns.pairplot(data_numeric,y_vars=['SalePrice'],x_vars=data_numeric.columns[i:i+5])
```

Relation between **SalePrice** and various ohter factors can clearly be observed. The plots with discrete values show a catagorical relationship with SalePrice. The plots with a downward slope show a negative relation whereas plots with upward slope show linear relationship. The density of different data points indicate what range the sale price falls maximum in with respect to a particular feature. For some features the relation is really obvious, like **GarageCars**, as the number increases, there can be seen a clear jump in price range. With 0, it was at max around 200000, with 2 it goes as high as 400000.

```python
#Plotting heatmap for correlation between various input features
corr = data_numeric.drop(['Id','SalePrice'], axis=1).corr()
fig, ax = plt.subplots(figsize=(20,20))  
sns.heatmap(corr[(corr>=0.5) | (corr<=-0.5)],
           cmap = 'viridis', vmax=1.0, vmin=-1, linewidths=0.1,
           annot=True, annot_kws={'size':10}, square=True,ax=ax)
```

The heatmap shows relation between various input parameters, which have relation higher than a particular value with each other. This helps select the features for the model so as to have a good inter-relationship which imporves the quality of prediction of the model.

```python
sns.regplot(data['GrLivArea'],data['SalePrice'])
```

The regression plot shows how much a particular input variable is related to the target variable. Here it can be seen, a the density of points b/w GrLivArea of 1000 to 3000 is really close to the line of regression, also there are a few extreme values around. In the range of 4000 to 5000 there can be seen some outliers.
